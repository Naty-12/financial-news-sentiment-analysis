```python
from data_loader import load_analyst_data, load_stock_data
from sentiment_analysis import compute_sentiment
from correlation_analysis import merge_data, compute_correlation, plot_correlation
```

```python
# Step 1: Load Data
df = load_analyst_data()
tickers = ['AAPL', 'AMZN','GOOG', 'META','MSFT','NVDA', 'TSLA']  # Example tickers
```

```python
# Step 2: Run analysis for each ticker
for ticker in tickers:
    stock_df = load_stock_data(ticker)
    daily_sentiment = compute_sentiment(df, ticker)
    merged_df = merge_data(daily_sentiment, stock_df)
    correlation = compute_correlation(merged_df)
    print(f"ðŸ“ˆ Correlation between sentiment and return for {ticker}: {correlation:.4f}")
    plot_correlation(merged_df, ticker)
```
